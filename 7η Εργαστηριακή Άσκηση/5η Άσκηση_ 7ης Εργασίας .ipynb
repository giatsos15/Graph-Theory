```python
from typing import List
import sys

SIZE = 4


class Queue:
    def __init__(self, capacity: int):
        self.cap = capacity
        self.front = self.size = 0
        self.near = capacity - 1
        self.arr = [None] * self.cap

    def is_full(self) -> bool:
        return self.size == self.cap

    def is_empty(self) -> bool:
        return self.size == 0

    def push(self, item: int):
        if self.is_full():
            print("queue is full!")
            return
        self.near = (self.near + 1) % self.cap
        self.arr[self.near] = item
        self.size += 1

    def pop(self) -> int:
        if self.is_empty():
            print("Empty queue!")
            return sys.maxsize
        item = self.arr[self.front]
        self.front = (self.front + 1) % self.cap
        self.size -= 1
        return item


def create_queue(capacity: int) -> Queue:
    return Queue(capacity)


def is_bipartite_check(G: List[List[int]], x: int, colorarr: List[int]) -> bool:
    colorarr[x] = 1
    queue = create_queue(SIZE)
    queue.push(x)

    while not queue.is_empty():
        x = queue.pop()
        if G[x][x] == 1:
            return False
        for i in range(SIZE):
            if G[x][i] and colorarr[i] == -1:
                colorarr[i] = 1 - colorarr[x]
                queue.push(i)
            elif G[x][i] and colorarr[i] == colorarr[x]:
                return False

    return True


def is_bipartite(G: List[List[int]]) -> bool:
    colorarr = [-1] * SIZE
    for i in range(SIZE):
        if colorarr[i] == -1:
            if not is_bipartite_check(G, i, colorarr):
                return False
    return True


if __name__ == "__main__":
    rows, cols = map(int, input("Write down below how many columns and rows has your graph.\nrows=: cols=:").split())
    G1 = [[0] * cols for _ in range(rows)]

    print("Enter your graph:")
    for i in range(rows):
        for j in range(cols):
            G1[i][j] = int(input(f"G1[{i}][{j}]="))

    print("\nGRAPH OUTPUT")
    for row in G1:
        print(*row)

    print("\nRESULTS")
    if is_bipartite(G1):
        print("This graph is 2-Chromatic because it is also bipartite Graph!")
    else:
        print("This is not a bipartite graph so it is not also a 2-Chromatic graph!")
```

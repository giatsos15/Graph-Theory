```python
import random
import networkx as nx
import matplotlib.pyplot as plt

# Main function 
def main():
    # Create a sequence of n (10 in this example) integers.
    sequence = degree_sequence(10)
    # If the degree sequence is graphical draw the graph or print an appropriate message.
    if (graphExists(sequence)):
        drawGraph(sequence)
    else:
        print("\nThe degree sequence is not graphical!") 

# Function that returns a list of n random integers. 
def degree_sequence(n):
    sequence = []
    a = n-1
    for i in range (0,n):
        # Every node of a graph can have a weight from 0 to n-1, where n is the order of the graph.
        n = random.randint(0,a)
        sequence.append(n)
        print(sequence)
    return sequence
 
# Function that returns true if a simple graph exists
def graphExists(s):
 
    # Continue parsing the code until one of the stopping conditions is met.
    while True:
 
        # Sort the list in a non-decreasing order
        s = sorted(s, reverse = True)
 
        # Check if all the elements of the list are equal to 0 and terminate if so. The sequence is graphical!
        if s[0]== 0 and s[len(s)-1]== 0:
            print("The degree sequence is graphical using the Havel-Hakimi algorithm!")
            return True
 
        # Store the first element of the list in a variable and delete it from the list.
        v = s[0]
        s = s[1:]
 
        # First, check if there are elements in the list as many as the first integer indicates. 
        # If not terminate, the sequence is not graphical.
        if v>len(s):
            return False
 
        # Then, subtract 1 from each of the next v elements.
        for i in range(v):
            s[i]-= 1
 
            # Last, if a negative element is encountered after the subtraction, 
            # terminate because the sequence is not graphical.
            if s[i]<0:
                return False

# Function to draw a graph.
def drawGraph(sequence):

    # Create empty graph
    G = nx.Graph()

    # Add nodes with weights from degree sequence
    for weight in sequence:
        G.add_node(weight)

    # Add edges based on degree sequence
    for i in range(len(sequence)):
        for j in range(i+1, len(sequence)):
            if sequence[i] > 0 and sequence[j] > 0:
                G.add_edge(i, j)
                sequence[i] -= 1
                sequence[j] -= 1

    # Draw graph
    print("Here is the Graph!")
    nx.draw(G, with_labels=True)
    plt.show()

if __name__ == "__main__":
    main()
```
